```python
from sklearn.model_selection import  train_test_split
```

```python
train,test = train_test_split(df['Close'],test_size=0.3 , shuffle=False)
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
Mmax_scaler = MinMaxScaler(feature_range=(0,1))

data_training_array = Mmax_scaler.fit_transform(train)
```

```python
data_training_array.shape[0]
```

```python
x_train=[] # this is list 
y_train=[]

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i,0])

x_train,y_train = np.array(x_train),np.array(y_train)
```

Model building 


```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()

model.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))


model.add(LSTM(units=60, activation='relu', return_sequences=True))
model.add(Dropout(0.3))

model.add(LSTM(units=80, activation='relu', return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(units=120 , activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss= 'mean_squared_error')
model.fit(x_train,y_train , epochs=50)
```

```python
past_100_days = train.tail(100)
```

```python
final_df = pd.concat([past_100_days, test], ignore_index=True)
```

```python
input_data = Mmax_scaler.fit_transform(final_df)
```

```python
x_test = []
y_test = []

for i in range(100,input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i,0])

x_test,y_test = np.array(x_test),np.array(y_test)
```

```python
x_test.shape
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
Mmax_scaler.scale_
```

```python
scaler_factor = 1/0.00606575

y_pred = y_pred*scaler_factor

y_test = y_test*scaler_factor
```

```python
import numpy as np

# Flatten predictions
y_pred_flat = y_pred.reshape(-1)

plt.figure(figsize=(12,6))
plt.plot(y_test, label='Original Price', linewidth=1)
plt.plot(y_pred, label='Predicted Price', linewidth=1)
plt.legend()
plt.show()
```
